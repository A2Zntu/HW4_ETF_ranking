```python
import pandas as pd
import pandas as pd
import time
import time
from selenium import webdriver
driver = webdriver.Chrome("chromedriver.exe")

driver.get('https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yieldYear&year=2019')
```

```python
table = driver.find_element_by_class_name('t-chart')
```

```python
type(table)
```

```python
def get_row_count(self):
    return len(self.find_elements_by_tag_name('tr')) - 1

def get_column_count(self):
    return len(self.find_elements_by_xpath("//tr[2]/td"))

def get_table_size(self):
    return {"rows": self.get_row_count(),
            "columns": self.get_column_count()}
```

```python
def row_data(self, row_number):
    if(row_number == 0):
        raise Exception("Row number starts from 1")

    row_number = row_number + 1;
    row = self.find_elements_by_xpath("//tr["+str(row_number)+"]/td");
    rData = [];
    for webElement in row :
        rData.append(webElement.text);

    return rData;
```

```python
pocket = pd.DataFrame(row_data(table, 1))

for i in range(2,get_row_count(table)+1):
    temp = row_data(table,i)
    tempdf = pd.DataFrame({str(i-1):temp})
    pocket = pd.concat([pocket, tempdf], axis=1)

pocket
```

```python
pocket_transposed = pocket.T
pocket_transposed = pocket_transposed.set_index([0])
```
