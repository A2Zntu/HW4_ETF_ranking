```python
import pandas as pd
import numpy as np
import scipy.optimize
import math
import os
```

# Read file (monthly)

```python
work_dir = os.getcwd()
df = pd.read_csv(os.path.join(work_dir, "data", "output_monthly_with_rf.csv"), index_col = False)
df
```

# Calculate excess return

```python
clean_data = df.iloc[1:, ]
d = pd.DataFrame()

colnames = list(clean_data.columns.values)[1:-1]
for col in colnames:
    d[col] = clean_data[col]-clean_data['Rf']
    ret_list = d[col].dropna().values.tolist()
    if np.mean(ret_list)<0:
        print(col, "negative expected return")
    
d
```

# Calculate riskiness r

```python
r =list()
```

```python
for col in colnames:
    ret_list = d[col].dropna().values.tolist()
    initial_r = np.var(ret_list) / (2 * np.mean(ret_list)) 
    def F(x):
        newlist = [-n / x for n in ret_list]
        return np.mean(np.exp(newlist))-1
    try:
        x = scipy.optimize.excitingmixing(F,initial_r)
        r.append(np.asscalar(x)) 
    except:
        r.append('nan')
```

```python
r
```

# Turn riskiness into Q

```python
Q_list = list()
for i in r:
    if i!='nan':
        temp = np.exp(-(1/i))
        Q_list.append(temp) 
    else:
        Q_list.append("nan")
Q_list
```

```python
table = pd.DataFrame(list(zip(Q_list, colnames)), 
               columns =['Q', 'ETF']) 
table
```

```python
sorted_table = table.sort_values(by=['Q'])
sorted_table['Ranking'] = np.arange(len(sorted_table))+1
riskiness_m = sorted_table
print (riskiness_m)
```

# Now, repeat the same procedure for weekly data

```python
work_dir = os.getcwd()
df = pd.read_csv(os.path.join(work_dir, "data", "output_weekly_with_rf.csv"), index_col = False)
```

```python
clean_data = df.iloc[1:, ]
d_w = pd.DataFrame()

colnames = list(clean_data.columns.values)[1:-1]
for col in colnames:
    d_w[col] = clean_data[col]-clean_data['Rf']
    ret_list = d_w[col].dropna().values.tolist()
    if np.mean(ret_list)<0:
        print(col, "negative expected return")
    
d_w
```

```python
r_w =list()
```

```python
for col in colnames:
    ret_list = d_w[col].dropna().values.tolist()
    initial_r = np.var(ret_list) / (2 * np.mean(ret_list)) 
    
    def F(x):
        newlist = [-n / x for n in ret_list]
        return np.mean(np.exp(newlist))-1
    try:
        x = scipy.optimize.excitingmixing(F,initial_r)
        r_w.append(np.asscalar(x)) 
    except:
        r_w.append('nan')
```

```python
Q_list = list()
for i in r_w:
    if i!='nan':
        temp = np.exp(-(1/i))
        Q_list.append(temp) 
    else:
        Q_list.append("nan")
Q_list

table = pd.DataFrame(list(zip(Q_list, colnames)), 
               columns =['Q', 'ETF']) 
table = table[table['Q'] != 'nan']
sorted_table = table.sort_values(by=['Q'])
sorted_table['Ranking'] = np.arange(len(sorted_table))+1
riskiness_w = sorted_table
print (riskiness_w)
```
