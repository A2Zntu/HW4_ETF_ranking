```python
import pandas as pd
import numpy as np
import scipy.optimize
import math
```

```python
df = pd.read_csv('testoutput_m.csv')
df
```

# Calculate excess return

```python
d = df
colnames = list(d.columns.values)[2:]
for col in colnames:
    d[col] = d[col]-d['rf']

d = d.drop(['rf'], axis=1)    
d
```

# Calculate riskiness r

```python
colnames = list(d.columns.values)[1:]

r = list()

for col in colnames:
    print(col)
    ret_list = d[col].dropna().values.tolist()
    initial_r = np.var(ret_list) / np.mean(ret_list)
    
    def F(x):
        newlist = [-n / x for n in ret_list]
        return np.mean(np.exp(newlist))-1
    try:
        x = scipy.optimize.excitingmixing(F,initial_r)
        r.append(np.asscalar(x)) 
    except:
        r.append('nan')
```

```python
r
```

# Turn riskiness into Q

```python
Q_list = list()
for i in r:
    if i!='nan':
        temp = np.exp(-(1/i))
        Q_list.append(temp) 
    else:
        Q_list.append("nan")
Q_list
```

```python
table = pd.DataFrame(list(zip(Q_list, colnames)), 
               columns =['Q', 'ETF']) 
table = table[table['Q'] != 'nan']
sorted_table = table.sort_values(by=['Q'])
sorted_table['Ranking'] = np.arange(len(sorted_table))+1
print (sorted_table)
```
